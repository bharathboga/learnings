## Indexing, Slicing, Boolean masking and indexing

```python
ra = np.random.randint(2,10, size = (4,4))
print(ra)
```

```python
#indexing and slicing
print(ra[0]) #row index 0
print(ra[1,2]) #element at row index 1 and column index 2
print(ra[0:2]) # rows from index 0 to index 2(excluded)
print(ra[:,[1,3]]) #all rows but of column index 1 and 3
print(ra[1:,[0,1]]) # all rows from index 1, column index 0 and 1
print(ra[([0,1],[1,3])])
#we can modify too
print(ra)
ra[:,[1,3]] = 0
print(ra)

# if ra is one d array
# ra[index position(s)]

# if ra is two d array
# ra[which row(s), which column(s)]

# if ra is three d array
# ra[which array(s), which row(s), which column(s)]

# if ra is four d array
# four d can be thought of boxes where each box consists of a set of arrays
```

```python
#accessing diagonal elements
print(ra[range(4),range(4)])
```

#### np.all and np.any

```python
ar = np.random.randint(3,10,size = (2,5))
print(ar)
```

```python
#np.all: Test whether all array elements along a given axis evaluate to True.
print(np.all(ar > 4,axis = 0)) #column wise
print(np.all(ar > 4,axis = 1)) #row wise
print(np.all(ar > 4,axis = None)) # logical AND over all the dimensions
```

```python
#np.any: Test whether any array elements along a given axis evaluate to True.
print(np.any(ar % 4 == 0,axis = 0)) #column wise
print(np.any(ar % 4 == 0,axis = 1)) #row wise
print(np.any(ar % 4 == 0,axis = None)) # logical OR over all the dimensions
```

## nditer a more sophisticated way of iterating through an array

```python
ar = np.random.randint(0,10, size = (4,4))
print(ar)
```

```python
for x in np.nditer(ar,order = 'F'): print(x,end = " ")
print()
for x in np.nditer(ar,order = "C"): print(x, end = " ")
print()
for x in np.nditer(ar, order = 'F',flags = ['external_loop']): print(x)
print()
for x in np.nditer(ar, op_flags = ['readwrite']):  x[...] = x*x
print(ar)
# ... triple dot is called ellipsis 
```

### Iterating through two arrays

```python
a = np.random.randint(0,10,size = (3,4))
b = np.random.randint(0,10,size = (3,1))
print(a)
print(b)
#we are broadcasting here..
#two arrays can be broadcastable iff the dimensions of the two arrays are compatible
#the two dimensions are compatible if and only if 1) dimensions are same or 2) one of them is 1
for x,y in np.nditer([a,b]):
    print(x,y,end = " / ")
```

#### when compared to python lists numpy arrays are fast, convenient, memory efficient.

```python
import numpy as np
import sys
```

```python
#memory
l = list(range(100))
print(f"python list size: {sys.getsizeof(int) * len(l)} bytes")
print(f"ar[0:20]: {l[0:20]}")

ar = np.arange(0,100)
print(f"np array size: {ar.nbytes} bytes")
print(f"ar[0:20]: {ar[0:20]}")

#we can change the datatype as desired and save memory. (using astype)
ar = ar.astype(np.uint8)
print(f"np array size: {ar.nbytes} bytes")
print(f"ar[0:20]: {ar[0:20]}")
```
